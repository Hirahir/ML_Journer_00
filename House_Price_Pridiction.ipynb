```python
# Step 1: Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 2: Create sample dataset
data = {
    'size_in_sqft': [500, 1000, 1500, 2000, 2500],
    'price_in_lakhs': [15, 30, 45, 60, 75]
}
df = pd.DataFrame(data)

# Step 3: Train model
X = df[['size_in_sqft']]
y = df['price_in_lakhs']
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict new price
print("Prediction for 1800 sqft:", model.predict([[1800]]))

# Step 5: Visualize
plt.scatter(df['size_in_sqft'], df['price_in_lakhs'], color='blue')
plt.plot(df['size_in_sqft'], model.predict(X), color='red')
plt.xlabel("Size (sqft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction")
plt.show()
```
